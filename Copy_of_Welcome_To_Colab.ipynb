```
import pandas as pd
import matplotlib.pyplot as plt
sp=pd.read_csv('/content/mtcars.csv')
auto=(sp['am']==0).sum()
manual=(sp['am']==1).sum()

plt.bar(['Automatic', 'Manual'], [auto, manual], color=['orange', 'blue'])
plt.title("Frequency of Transmission Types")
plt.ylabel("Number of Cars")
plt.grid(axis='y')
plt.show()
```

```
plt.scatter(sp['wt'], sp['mpg'], color='green')
plt.title("Car Weight vs MPG")
plt.xlabel("Weight (1000 lbs)")
plt.ylabel("Miles per Gallon")
plt.grid(True)
plt.show()
```

```
plt.boxplot(sp['mpg'])
plt.title("Box Plot of MPG")
plt.ylabel("Miles per Gallon")
plt.grid(True)
plt.show()
```
